Q 02-01 - Escreva um programa que imprima a soma dos números ÍMPARES de 1 a 100. A lógica do programa é representada pelo diagrama de fluxo abaixo.

image.png

````python
# Inicializa a variável n com valor 1, que será o primeiro número ímpar a ser somado.
n = 1
# Inicializa a variável s com valor 0, que armazenará a soma dos números ímpares.
s = 0

# Inicia um loop while que continuará enquanto n for menor ou igual a 100.
while n <= 100:
    # A cada iteração, o valor atual de n (número ímpar) é adicionado à soma s.
    s = s + n
    # O valor de n é incrementado em 2 para garantir que o próximo número seja ímpar.
    n = n + 2

# Exibe o resultado da soma dos números ímpares de 1 a 100.
print('A soma dos números ímpares de 1 a 100:', s)
````

Q 02-02 - Escreva um programa que recebe um número inteiro positivo do usuário e diz se é um número par ou ímpar. Use o pseudo-código para se guiar:

```
print 'Digite um número inteiro positivo:'
n = input()
se resto da divisão por 2 é igual a 0:
    print 'n é par'
se não:
    print 'n é ímpar'
```



````python
# Exibe a mensagem para o usuário digitar um número inteiro positivo.
print('Digite um número inteiro positivo:')

# A função input() recebe a entrada do usuário como string.
# A função int() converte a string recebida em um número inteiro e armazena na variável n.
n = int(input())

# Verifica se o número é par usando o operador de módulo (%).
# O módulo retorna o resto da divisão de n por 2.
# Se o resto for 0, então o número é divisível por 2 e, portanto, é par.
if n % 2 == 0:
    # Se a condição for verdadeira, imprime que o número é par.
    print(n, 'é par')
else:
    # Caso contrário, imprime que o número é ímpar.
    print(n, 'é impar')
````

Q 05-01 Usando os operadores lógicos "and", "or" e "not" escreva um programa que, dado x = 1 e y = 0, imprima os seguintes resultados:


0

1

False

True

````python
# Atribui o valor 1 à variável x.
x = 1

# Atribui o valor 0 à variável y.
y = 0

# O operador 'and' retorna True (ou 1) se ambos os operandos forem verdadeiros (ou diferentes de 0).
# Caso contrário, retorna o primeiro valor falso encontrado. Como y é 0 (falso), o resultado será 0.
print(x and y)  # Resultado: 0

# O operador 'or' retorna True (ou 1) se pelo menos um dos operandos for verdadeiro.
# Caso x seja verdadeiro (diferente de 0), ele será retornado sem verificar y. Como x é 1, o resultado será 1.
print(x or y)  # Resultado: 1

# O operador 'not' inverte o valor lógico da variável.
# Se x é 1 (verdadeiro), 'not x' será False (ou seja, 0).
print(not x)  # Resultado: False (ou 0)

# O operador 'not' aplicado a y, que é 0 (falso), retornará True (ou seja, 1).
print(not y)  # Resultado: True (ou 1)
````

Q 06-01 - Escreva um programa que leia 2 entradas do usuário e as imprima do menor para o maior.

(Restrição: dois números inteiros não são o mesmo número)

Exemplo:


Insira dois números inteiros: 14 10
Os números inseridos em ordem crescente são: 10 14

````python
# A função input() solicita ao usuário que digite um número inteiro.
# O valor digitado é convertido para um número inteiro com a função int() e armazenado na variável 'numero1'.
numero1 = int(input('Digite um número inteiro: '))

# A função input() solicita ao usuário que digite outro número inteiro.
# O valor digitado é convertido para um número inteiro com a função int() e armazenado na variável 'numero2'.
numero2 = int(input('Digite um outro número inteiro: '))

# O if verifica se 'numero1' é menor que 'numero2'.
# Se a condição for verdadeira, imprime os dois números na ordem original (numero1, numero2).
if numero1 < numero2:
    print('Os números inseridos em ordem crescente são:', numero1, numero2)
# Caso contrário (se numero1 for maior ou igual a numero2), imprime os dois números invertidos,
# para que fiquem em ordem crescente (numero2, numero1).
else:
    print('Os números inseridos em ordem crescente são:', numero2, numero1)
````

Q 07-01 - Escreva um programa onde o usuário deve responder duas perguntas, se é maior de idade e se é casado, depois imprima o resultado.

Exemplo:


Você é maior de idade? (digite 1 se sim, 0 se não): 1

Você é casado(a)? (digite 1 se sim, 0 se não): 0

Você é maior de idade e solteiro.

````python
# A função input() solicita ao usuário que informe se ele é maior de idade (1 para sim, 0 para não).
# O valor digitado é convertido para um número inteiro com a função int() e armazenado na variável 'idade'.
idade = int(input('Você é maior de idade? (digite 1 se sim, 0 se não):'))

# A função input() solicita ao usuário que informe se ele é casado (1 para sim, 0 para não).
# O valor digitado é convertido para um número inteiro com a função int() e armazenado na variável 'casado'.
casado = int(input('Você é casado(a)? (digite 1 se sim, 0 se não):'))

# O if verifica se a variável 'idade' tem o valor 1 (ou seja, o usuário é maior de idade).
# Se for verdadeiro (idade == 1), imprime 'Você é maior de idade e' sem quebrar a linha (usando o argumento 'end').
if idade == 1:
    print('Você é maior de idade e', end=' ')
# Caso contrário (idade == 0), imprime 'Você é menor de idade e' também sem quebrar a linha.
else:
    print('Você é menor de idade e', end=' ')

# O if verifica se a variável 'casado' tem o valor 1 (ou seja, o usuário é casado).
# Se for verdadeiro, imprime 'casado'.
if casado == 1:
    print('casado')
# Caso contrário (casado == 0), imprime 'solteiro'.
else:
    print('solteiro')
````

Q 08-01 - Entre os números naturais positivos diferentes de 1, um número que não é primo é chamado de número composto. Imprima os números primos e compostos de 2 a 12.

Uma expressão deve determinar se o número é primo ou composto, caso seja composto, deve ser impresso na tela.

````python
# Cria uma lista de números de 2 a 12 (inclusive) usando a função range().
numeros = list(range(2, 13))

# Inicia um loop que itera sobre os índices da lista 'numeros'.
for i in range(0, len(numeros)):
    # A variável 'num' armazena o número atual da lista 'numeros' que está sendo analisado.
    num = numeros[i]

    # Inicializa a variável 'soma' com 0, que será usada para contar quantos divisores o número 'num' possui.
    soma = 0

    # Inicia um loop que itera sobre os índices da lista 'numeros' novamente para verificar os divisores.
    for j in range(0, len(numeros)):
        # Verifica se 'num' é divisível pelo número na posição 'j' da lista 'numeros'.
        # Se o resto da divisão for 0, significa que 'num' é divisível por 'numeros[j]'.
        if num % numeros[j] == 0:
            # Incrementa a variável 'soma' em 1 para contar mais um divisor encontrado.
            soma = soma + 1

    # Após contar todos os divisores, verifica se 'soma' é igual a 1.
    # Um número é considerado primo se tiver exatamente 2 divisores: 1 e ele mesmo.
    # Como estamos começando a contagem a partir de 2, se 'soma' for 1, significa que o número é primo.
    if soma == 1:
        print('O número', num, 'é primo!')
    else:
        print('O número', num, 'é composto!')
````

Q 09-01 Um número Armstrong é um número inteiro de três dígitos que é igual à soma dos cubos de cada dígito. Encontre todos os números Armstrong entre números inteiros de três dígitos e imprima-os.

Exemplo:

153 é um número Armstrong, pois 1³+5³+3³ = 1+125+27 = 153

````python
# Cria uma lista de números de 100 a 999 (inclusive) usando a função range().
numeros = list(range(100, 1000))

# Inicia um loop que itera sobre cada número 'i' na lista 'numeros'.
for i in numeros:
    # Converte o número atual 'i' em uma string para acessar os dígitos individualmente.
    s = str(i)

    # Calcula a soma dos cubos dos dígitos do número.
    # O primeiro dígito é convertido de volta para inteiro, elevado ao cubo e somado.
    # O mesmo é feito para o segundo e terceiro dígitos.
    soma = (int(s[0]) * int(s[0]) * int(s[0])) + (int(s[1]) * int(s[1]) * int(s[1])) + (int(s[2]) * int(s[2]) * int(s[2]))

    # Verifica se a soma dos cubos dos dígitos é igual ao número original 'i'.
    if soma == i:
        # Se for verdadeiro, imprime que o número é um número Armstrong.
        print('O número', i, 'é um número Armstrong')
    else:
        # Caso contrário, imprime que o número não é um número Armstrong.
        print('O número', i, 'não é um número Armstrong')
````
